```python
# add a caption to image with AI CoCa（Contrastive Captioner）

# GPU or CPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# reading AI model
model, _, transform = open_clip.create_model_and_transforms(
    "coca_ViT-L-14",
    pretrained="mscoco_finetuned_laion2B-s13B-b90k",
    device=device,
)

filenames = []
for movie_file in sorted(
    glob.glob(os.path.join('movies', '*.mp4'))):
    print( movie_file )
    filenames.append( movie_file )

dt_now = datetime.datetime.now()
os.makedirs("./text/", exist_ok=True)
gen_sentence_out = './text/movie_time_caption_{}.txt'\
    .format( dt_now.strftime('%Y%m%d_%H%M%S'))
f = open(gen_sentence_out, 'w')        
for filename in filenames:        
    dir_path = filename.split( '.' )[0]
    dir_path = dir_path.replace( "movies", "images")        
    for img_file in sorted(
        glob.glob(os.path.join(dir_path, '*.jpg'))):
            
        # reading imge
        img = Image.open(img_file)
        #img = transforms(img)
        #img = img.unsqueeze(0)
        #img = img.to(config.device)


        ## disaplay image
        #img_plt = Image.open(img_file)
        #img_plt = img_plt.resize([224, 224], Image.LANCZOS)
        #plt.imshow(img_plt)
        #plt.axis('off')
        #plt.show()
        print(f'imput image: {os.path.basename(img_file)}')

        # 画像からキャプションを生成
        im = transform(img).unsqueeze(0).to(device)
        with torch.no_grad(), torch.cuda.amp.autocast():
            generated = model.generate(im, seq_len=40, generation_type='top_p')

        # キャプションを人間が読める文章に変換して表示
        caption = (
            open_clip.decode(generated[0].detach())
            .split("<end_of_text>")[0]
            .replace("<start_of_text>", "")
        )

        sentence = caption
        print(f'output caption: {sentence}')

        # 推定結果を書き込み
        img_file2 = img_file.split("\\")[2]
        second = img_file2.split( "_" )[2]
        second = second.rstrip( ".jpg")
        second = float(second)
        print(filename + "<sep>" + str(second) + "<sep>" + sentence, file=f)
f.close()
```

```python
# evaluate cosine_similarity of image and text with openai clip-vit-base-patch32 model
# and make 00_movie_search_{time}.txt for  python's searh server( sentence-transformers ) program.

import datetime
from PIL import Image
import requests
import torch

from transformers import CLIPProcessor, CLIPModel
from transformers import AutoTokenizer, AutoProcessor, CLIPModel

cos_sim = torch.nn.CosineSimilarity( dim = 1 )

model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
tokenizer = AutoTokenizer.from_pretrained("openai/clip-vit-base-patch32")
processor = AutoProcessor.from_pretrained("openai/clip-vit-base-patch32")

#dir1 = os.chdir("C:\\Users\\uchiyats\\python-study\\movie_search_for_english_report\\")


# reading image
cos_sim1 = []
for text, image1 in zip( texts, imgs):

    inputs = tokenizer(text, padding=True, return_tensors="pt")
    text_features = model.get_text_features(**inputs)

    image = Image.open(image1)
    inputs = processor(images=image, return_tensors="pt")
    image_features = model.get_image_features(**inputs)

    cos_sim0 = torch.squeeze(cos_sim( image_features, text_features ), dim = 0 ).item()
    cos_sim1.append( cos_sim0 )
    
    print( "image1:", image1, "text:", text,  "cos_sim0:", cos_sim0 )
    
#dir1 = os.chdir("C:\\Users\\uchiyats\\python-study\\movie_search_for_english_report\\")
dt_now = datetime.datetime.now()
#gen_sentence_out = './00_movie_search_{}.txt'\
#        .format( dt_now.strftime('%Y%m%d_%H%M%S'))
gen_sentence_out = './00_movie_search.txt'
f = open(gen_sentence_out, 'w')

for text, mv, sec, cos1 in zip( texts, movie, second, cos_sim1 ):
    if cos1 > 0.05:
        text = text.replace( "\n", "" )
        print(mv + "<sep>" + str(sec) + "<sep>" + text, file=f)
        
        
f.close()
```
